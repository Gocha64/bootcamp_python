```python
# 문제3

n, m = map(int,input().split())

info = []
for i in range(m):
  info.append(list(input().split()))

for i in info:
  i[0] = int(i[0])
  i[1] = int(i[1])


cm = 0
result = 0
tmp = True
red = set()
blue = set()
friendList = {}
enemyList = {}


for i in range(m):
  a, b, c = info[i]


  if i == 0:
    if info[i][2] == 'f':
      red.add(a)
      red.add(b)

    else:
      red.add(a)
      blue.add(b)

    continue

  if c == 'f':
    if a in red or b in red:
      red.add(b)
      red.add(a)
    elif a in blue or b in blue:
      blue.add(a)
      blue.add(b)
    else:
      friendList[a] = b
  else:
    if a in red or b in blue:
      red.add(a)
      blue.add(b)
    elif b in red or a in blue:
      blue.add(a)
      red.add(b)
    else:
      enemyList[a] = b

  

  for key, value in friendList.items():
    if key in red or value in red:
      red.add(key)
      red.add(value)
    elif key in blue or value in blue:
      blue.add(key)
      blue.add(value)

  
  for key, value in enemyList.items():
    if key in red or value in blue:
      red.add(key)
      blue.add(value)
    elif key in blue or value in red:
      blue.add(key)
      red.add(value)

  if blue.intersection(red) != set():
    if cm == 0:
      tmp = False
      result = i+1
      cm+= 1

if tmp == True:
  print("THE SPY IS LOYAL")
else:
  print(result)


```
