```python
# 문제 1

n = int(input(""))

if n > 1000:
  print("1000이하의 수를 입력하시오")

else:
  total = 0

  for i in range(1, n):
    if i % 3 == 0 or i % 5 == 0:
      #print(i)
      total += i
  print(total)
```
